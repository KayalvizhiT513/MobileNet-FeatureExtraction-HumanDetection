```python
import os
import glob
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import cv2
```

```python
i = 0
for img_path in glob.glob(os.path.join("/content/drive/MyDrive/datasets/JPEGImages","*.jpg")):
  image = cv2.imread(img_path)

  with open(os.path.join("/content/drive/MyDrive/datasets/Annotations",(os.path.basename(img_path).replace("jpg","xml"))))as f:

    data = f.read()
    parsedData = BeautifulSoup(data)
    objects = parsedData.find_all("object")


    for obj in objects:

      xmin = int(obj.find("bndbox").find("xmin").text)
      ymin = int(obj.find("bndbox").find("ymin").text)
      xmax = int(obj.find("bndbox").find("xmax").text)
      ymax = int(obj.find("bndbox").find("ymax").text)


      xmin = max(0,xmin)
      ymin = max(0,ymin)
      xmax = min(image.shape[1],xmax)
      ymax = min(image.shape[0],ymax)



      if(xmax>xmin and ymax>ymin):
        crop = image[ymin:ymax, xmin:xmax]
        cv2.imwrite(os.path.join("/content/drive/MyDrive/datasets/PositiveSamples1000",str(i)+".jpg"),crop)
        i+=1
        print("Generated Image:",i)

      else:
        print("Invald BoudingBox")
  if(i>=1000):
      break
```
